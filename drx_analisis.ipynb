```python
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, peak_widths
from scipy.optimize import curve_fit


```

```python
drxs={}

numero_de_muestras=10

for i in range(numero_de_muestras):
    numero_de_paginas=i
    drxs[f'drx{i}']=pd.read_excel('ruta/drx_urea_excel.xlsx', sheet_name=numero_de_paginas)
    drxs[f'drx{i}'] = drxs[f'drx{i}'][drxs[f'drx{i}'].iloc[:, 0] <= 71.00]
```

```python
drxs['drx0'].info()
```

```python

data = {}
picos = {}

K=0.9
L=0.154187

for i in range(numero_de_muestras):
    x = drxs[f'drx{i}'].iloc[:, 0]
    y = drxs[f'drx{i}'].iloc[:, 1]

    # Encontrar los picos en el patrón de difracción
    peaks, _ = find_peaks(y, height=1000, width=15)
    picos_x = x[peaks]
    intensidades = y[peaks]
    results_half = peak_widths(y, peaks, rel_height=0.5)

    fwhm_values = results_half[0]

    # Crear diccionario con la información de los picos de drx
    data[f'drx{i}'] = {'posicion_del_pico': picos_x,
                       'intensidad_del_pico': intensidades,
                       'FWHM': fwhm_values}

    # Crear DataFrame con la información recopilada
    picos[f'df{i}'] = pd.DataFrame(data[f'drx{i}'])

    angulo_theta_grados = (picos[f'df{i}']['posicion_del_pico']) / 2
    angulo_theta_grados = angulo_theta_grados.values.tolist()

    # Convertir a radianes utilizando numpy.radians()
    angulo_theta_radianes = np.radians(angulo_theta_grados)

    # Calcular coseno de theta
    cos_theta = np.cos(angulo_theta_radianes)

    # Convertir FWHM a radianes
    fwhm_rad = np.radians(picos[f'df{i}']['FWHM'])

    # Ecuación de scherrer

    
    B = (K*L)/((fwhm_rad)*(cos_theta))

    for i, cristalita in enumerate(B):
        print(f"El valor de la cristalita para el pico {i+1}: es {cristalita}")

    print('######################################################################################################################################')
    

    # Imprimir la posición de los picos
    for j, pico in enumerate(picos_x):
        print(f"Posición del pico {j+1}: {pico}")

    print('######################################################################################################################################')

    # Imprimir los valores de FWHM
    for j, fwhm in enumerate(fwhm_values):
        print(f'FWHM del pico {j+1}: {fwhm}')

    
    plt.plot(x, y)
    plt.plot(picos_x, intensidades, "x")
    plt.show()
```
