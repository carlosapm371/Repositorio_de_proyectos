```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot  as plt 
import seaborn as sns 
from bs4 import  BeautifulSoup
import requests 
import re 
import html5lib
from html5lib import parse 
```

```python
website = 'https://www.nba.com/lakers/history/season-by-season-records'
respuesta = requests.get(website)
contenido = respuesta.text
soup = BeautifulSoup(contenido, 'html5lib')
rows = soup.find_all('tr')

print(len(rows))  # Muestra el número de divs en la página
```

```python
nombres_columnas =['Season', 'W', 'L','W-L%', 'Finished', 'Playoffs', 'Coaches']
```

```python
results = []

for row in rows:
   
    row_data = []

    cells = row.find_all('td')

    for cell in cells:
        row_data.append(cell.get_text())

        results.append(row_data)


df = pd.DataFrame(results, columns=nombres_columnas)


df
```
