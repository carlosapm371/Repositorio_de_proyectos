```python
import pandas as pd 
import numpy as numpy 
import matplotlib.pyplot as plt 
import seaborn as sns
from bs4 import BeautifulSoup
import requests
import re 
import html5lib
from html5lib import parse 
```

```python
datos = []

website = 'https://www.luckynumber.mx/melate-retro-resultados'
sitios = [website]

for i in  range(2, 14):
    nombre_sitio='https://www.luckynumber.mx/melate-retro-resultados?page=' + str(i)
    sitios.append(nombre_sitio)

```

```python
sitios
```

```python
for sitio in sitios:
    respuesta = requests.get(sitio)
    contenido_html = respuesta.text
    soup = BeautifulSoup(contenido_html, 'html5lib')
    rows = soup.find_all('tr')

    for row in rows: 
        row_data=[]
        cells = row.find_all('td')

        for cell in cells:
            row_data.append(cell.get_text())

        datos.append(row_data)    
```

```python
datos
```

```python
nombres_columnas = ['Sorteo', 'Fecha', 'Numero_1', 'Numero_2', 'Numero_3', 'Numero_4', 'Numero_5', 'Numero_6', 'Numero_7', 'Bolsa']
```

```python
df = pd.DataFrame(datos, columns=nombres_columnas)
df=df.drop(index=0)
df['Fecha']=pd.to_datetime(df['Fecha'], format='%d/%m/%Y')
df=df.sort_values(by='Fecha', ascending=False)
df=df.dropna(axis=0, how='any')
#df.to_csv('/Users/carlosabraham1/Desktop/Historico_datos_melate.csv', index=False)
```

```python
columnas_float = [col for col in df.columns if df[col].dtype != 'datetime64[ns]']

for col in columnas_float:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print(df.dtypes)    
```
